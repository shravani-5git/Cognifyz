```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv")
print("Rows:", len(df))
df.head()
```

```python
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')

print("Converted Columns Summary:")
print(df[['Aggregate rating', 'Votes']].describe())
```

```python
bins = [0, 1, 2, 3, 4, 5]
labels = ["01", "12", "23", "34", "45"]

df["Rating Range"] = pd.cut(
    df["Aggregate rating"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

df[["Aggregate rating", "Rating Range"]].head()
```

```python
rating_distribution = df["Rating Range"].value_counts().sort_index()
print("Rating Distribution:")
rating_distribution
```

```python
most_common_range = rating_distribution.idxmax()
most_common_range
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x=rating_distribution.index,
            y=rating_distribution.values,
            palette="viridis")

plt.title("Distribution of Restaurant Rating Ranges")
plt.xlabel("Rating Range")
plt.ylabel("Number of Restaurants")
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.histplot(
    df["Aggregate rating"],
    bins=20,
    kde=True
)

plt.title("Histogram of Aggregate Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

```python
avg_votes = df["Votes"].mean()
min_votes = df["Votes"].min()
max_votes = df["Votes"].max()

print("Average number of votes received by restaurants:", round(avg_votes, 2))
print("Minimum votes:", min_votes)
print("Maximum votes:", max_votes)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Dataset .csv")
print("Rows:", len(df))
df.head()
```

```python
df['Cuisine Combo'] = (
    df['Cuisines']
    .str.split(',')
    .apply(lambda x: ", ".join(sorted([i.strip() for i in x if i.strip() != ""])))
)
```

```python
combo_counts = df['Cuisine Combo'].value_counts().head(10)
combo_counts
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=combo_counts.values, y=combo_counts.index, palette="viridis")
plt.title("Top 10 Most Common Cuisine Combinations")
plt.xlabel("Count")
plt.ylabel("Cuisine Combination")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
combo_rating = df.groupby('Cuisine Combo')['Aggregate rating'].mean().sort_values(ascending=False)

combo_rating.head(10)
```

```python
top_rated_combos = combo_rating.head(10)
plt.figure(figsize=(10,5))
sns.barplot(
    x=top_rated_combos.values,
    y=top_rated_combos.index,
    hue=top_rated_combos.index,
    palette="coolwarm",
    legend=False
)

plt.title("Top 10 Highest Rated Cuisine Combinations")
plt.xlabel("Average Rating")
plt.ylabel("Cuisine Combination")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Dataset .csv")
```

```python
lat_col = "Latitude"
lon_col = "Longitude"

df[lat_col] = pd.to_numeric(df[lat_col], errors='coerce')
df[lon_col] = pd.to_numeric(df[lon_col], errors='coerce')

geo_df = df.dropna(subset=[lat_col, lon_col])
print("Rows with valid coordinates:", len(geo_df))
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    geo_df[lon_col],
    geo_df[lat_col],
    s=8,
    alpha=0.3
)

plt.title("Restaurant Locations (Latitude vs Longitude)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hexbin(
    geo_df[lon_col],
    geo_df[lat_col],
    gridsize=40,
    cmap="viridis",
    mincnt=1
)

plt.title("Restaurant Density Map (Hexbin Clusters)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="Number of Restaurants")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(
    x=geo_df[lon_col],
    y=geo_df[lat_col],
    cmap="plasma",
    fill=True,
    bw_adjust=0.5
)

plt.title("Restaurant Heatmap (KDE Density)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Dataset .csv")
```

```python
name_col = "Restaurant Name"
rating_col = "Aggregate rating"
votes_col = "Votes"

df[rating_col] = pd.to_numeric(df[rating_col], errors="coerce")
df[votes_col] = pd.to_numeric(df[votes_col], errors="coerce")
```

```python
name_counts = df[name_col].value_counts()
chains = name_counts[name_counts > 1]
```

```python
print("Number of chains found:", len(chains))
chains.head(20)
```

```python
plt.figure(figsize=(10, 6))

top_chains = chains.head(10)

sns.barplot(
    x=top_chains.values,
    y=top_chains.index,
    hue=top_chains.index,
    palette="viridis",
    legend=False
)

plt.title("Top 15 Most Common Restaurant Chains")
plt.xlabel("Number of Outlets")
plt.ylabel("Restaurant Chain")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
chain_names = chains.index.tolist()
chain_df = df[df[name_col].isin(chain_names)]
```

```python
chain_stats = chain_df.groupby(name_col).agg(
    Locations = (name_col, 'count'),
    Avg_Rating = (rating_col, 'mean'),
    Total_Votes = (votes_col, 'sum')
).sort_values("Locations", ascending=False)

chain_stats.head(10)
```

```python
top10_by_rating = chain_stats.sort_values("Avg_Rating", ascending=False).head(10)

plt.figure(figsize=(10, 6))

sns.barplot(
    x=top10_by_rating["Avg_Rating"],
    y=top10_by_rating.index,
    hue=top10_by_rating.index,
    palette="viridis",
    legend=False
)

plt.title("Top 10 Restaurant Chains by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Restaurant Chain")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
top10_by_votes = chain_stats.sort_values("Total_Votes", ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(
    x=top10_by_votes["Total_Votes"],
    y=top10_by_votes.index,
    hue=top10_by_votes.index,
    palette="magma",
    legend=False
)

plt.title("Top 10 Restaurant Chains by Total Votes (Popularity)")
plt.xlabel("Total Votes")
plt.ylabel("Restaurant Chain")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
